```python
#unique values in all the columns
for col in df1:
    print(col)
    print(df1[col].unique())
    print('\n')
```

```python
sales=df1.Global_Sales.mean()
print(f"Mean Global sales=", sales)

sales=df1.EU_Sales.mean()
print(f"Mean sales in Europe=", sales)

sales=df1.NA_Sales.mean()
print(f"Mean sales in North America=", sales)

sales=df1.JP_Sales.mean()
print(f"Mean sales in Japan=", sales)
```

```python
P=df1.groupby('Platform')
P.get_group('Wii')
```

```python
# Yearly sales in North America, Europe, and Japan
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))    #width, height
sns.lineplot(x='Year', y='NA_Sales', data=df1)
plt.title('Sales in North America by Year')
plt.xlabel('Year')
plt.ylabel('Sales in North America')
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='EU_Sales', data=df1)
plt.title('Sales in Europe by Year')
plt.xlabel('Year')
plt.ylabel('Sales in Europe')
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='JP_Sales', data=df1)
plt.title('Sales in Japan by Year')
plt.xlabel('Year')
plt.ylabel('Sales in Japan')
plt.show()
```

```python
# Genre published by year
plt.figure(figsize=(14, 10))

sns.boxplot(x='Genre', y='Year', data=df1)
plt.xticks(rotation=90)
plt.grid()
plt.title('Genre Published by Year')
plt.xlabel('Genre')
plt.ylabel('Year')
plt.show()
```
