```python
from groq import Groq

import os
from dotenv import load_dotenv

load_dotenv()

groq_api_key = os.getenv("GROQ_API_KEY")

client = Groq(api_key=groq_api_key)
```

```python
prompt = "Classify iris flowers"

system_prompt = [
    {
        "role": "system",
        "content": "You are capable of assisting with classification tasks. You can classify any types of data, including text, images, and audio."
    }
]

user_prompt = [
    {
        "role": "user",
        "content": prompt
    }
]

system_prompt.extend(user_prompt)

llm_response = client.chat.completions.create(
    model= "llama3-70b-8192",
    messages = system_prompt,
    max_tokens = 500,
    temperature=1.5    
)

ai_response = llm_response.choices[0].message.content
print("Response: ", ai_response)
```

```python
system_prompt ={
        "role": "system",
        "content": "You are ChatMate, a helpful, friendly, and intelligent AI assistant designed to chat with users."
        "You are capable of assisting with classification tasks. You can classify any types of data, including text, images, and audio."
        "You have a vast knowledge of various topics from study materials to research papers to technical documentation."
        "You can also assist with coding tasks, including writing code snippets, debugging, and providing explanations."
        "You can also offer different coding suggestions to help the users."
        "Always begin with Hello, how can I assist you today?"
        "And, begin reply with 'AI: Sure! I can help you with that.'"
        "Don't repeat the content response from the user."
        "If the user asks for one response just give one response."
    }


chat_history = [system_prompt]

while True:
    user_input = input("User: ")
    
    if user_input.lower() == "exit":
        print("Goodbye! Take care.")
        break
    
    user_prompt = {
        "role": "user",
        "content": user_input
    }
    
    chat_history.append(user_prompt)
    
    llm_response = client.chat.completions.create(
        model= "llama3-70b-8192",
        messages = chat_history,
        max_tokens = 500,
        temperature=1.5    
    )
    ai_response = llm_response.choices[0].message.content
    
    ai_response_context = {
        "role": "assistant",
        "content": ai_response
    }
    chat_history.append(ai_response_context)
    print("AI: ", ai_response)

```

```python
import os
from dotenv import load_dotenv
load_dotenv()

GROQ_API_KEY = os.getenv("GROQ_API_KEY")
SERPAPI_KEY = os.getenv("SERPAPI_KEY")
```

```python
from langchain_groq import ChatGroq
from langchain_community.utilities import SerpAPIWrapper
from langchain.tools import Tool
from langchain.memory import ConversationBufferMemory
from langchain.agents import initialize_agent

groq_llm = ChatGroq(
    api_key=GROQ_API_KEY,
    model="llama3-70b-8192",
    temperature=1.5,
    max_tokens=500
)
search = SerpAPIWrapper(serpapi_api_key=SERPAPI_KEY)

search_tool = Tool(
    name="Web Search",
    func=search.run,
    description="Searches the web for the latest information.",
    handle_prasing_errors=True
)

memory = ConversationBufferMemory(memory_key = "history")
```

```python
agent = initialize_agent(
    tools=[search_tool],
    llm=groq_llm,
    memory = memory,
    verbose=True
)
```

```python
response = agent.invoke("Who is the current elected prime minister of Nepal?")
print("Response: ", response)
```
